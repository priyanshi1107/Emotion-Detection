# import all lib

```python
import matplotlib.pyplot as plt
import cv2
```

# imread()

```python
img = cv2.imread("/content/dog_backpack.png")#"00-puppy.jpg"
```

```python
img.shape
```

# imshow()

```python
plt.imshow(img)
```

```python
#matplotlib RGB
#OPENCV BGR
```

# cvtColor()

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

# Resize img

```python
rgb_img.shape
```

```python
new_img=cv2.resize(rgb_img,(1000,1000))
```

```python
plt.imshow(new_img)
```

# Flip img

```python
flip_img=cv2.flip(rgb_img,0)#0,1,-1
plt.imshow(flip_img)
```

```python
flip_img=cv2.flip(rgb_img,1)#0,1,-1
plt.imshow(flip_img)
```

```python
flip_img=cv2.flip(rgb_img,-1)#0,1,-1
plt.imshow(flip_img)
```

# imwrite()

```python
cv2.imwrite("_new.jpg",flip_img)
```

# large img

```python
plt.figure(figsize=(15,10))
plt.imshow(rgb_img)
```
